```python
# 네이버 검색 API 예제 - 블로그 검색
import os 
import sys
import urllib.request
import json
client_id='9bLbOBlav89U3DdVt3iO'
client_secret='VGEXL3fXtT'
encText = urllib.parse.quote("가챠")
url = "https://openapi.naver.com/v1/search/news?query=" + encText + "&display=100"
# 100가지 추출
# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
# 네이버에서는 헤더필요
response = urllib.request.urlopen(request)
# 요청이 정상적이라면 
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    res_body = json.loads(response_body.decode('utf-8')) # 한글 디코드 utf 필요함
    print(res_body['items'])
    str1=""
    
    # 특정 카테고리(title),(description) 추출
    for item in res_body['items']:
        str1+=item['title']+" "
        print(item['title'])
        print(item['description'])
else:
    print("Error Code:" + rescode)
```

```python
news=kkma.nouns(str1)
news
```

```python
# 차트화를 위한 텍스트
news_tx=Text(news)
# 빈도수 구해서 워드 클라우드
often=news_tx.vocab()
```

```python
# test 출력
from wordcloud import WordCloud
font_path='c:/Windows/fonts/malgun.ttf'
wc=WordCloud(width=10000,height=800, background_color='white',random_state=0,
             font_path=font_path)
plt.imshow(wc.generate_from_frequencies(often))
plt.axis('off')
plt.show()
```
