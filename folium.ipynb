```python
# ! pip install folium
import folium 
from folium import plugins 
```

```python
m=folium.Map(location=[34.8799083,128.7384361],zoom_start=13,
             tiles='Stamen Toner')
# m.save('data/map1.html')
m
```

location : 지도의 중심좌표(위도, 경도)
zoom_start:지도의 확대 정도 (1~18,default:10)
tiles: 지도 타일 : 'Stamen Terrain'(지형),'Open StreetMap'(로드맵),'Stamen Toner'(흑백)

```python
# 서울 시청의 위도, 경도
latitude=37.56667
longitude= 126.97806

# 덕수궁의 위도, 경도
latitude2=37.5658859
longitude2= 126.9754788

m=folium.Map(location=[latitude,longitude],zoom_start=16)
folium.Marker([latitude,longitude],popup='서울시청',icon=folium.Icon(color='red',icon_color='yellow',icon='info-sign')).add_to(m)
folium.Marker([latitude2,longitude2],popup='덕수궁').add_to(m)
m

```
