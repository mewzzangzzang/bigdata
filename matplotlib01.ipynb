데이터//

//
날짜와 최고기온, 최저기온을 가지고 그래프그리기

```python

filepath='data/temperature_2014.csv'
df=pd.read_csv(filepath)
# df.info()
# 데이터 가지고 오면 항상 타입이나 기타정보 확인하기
df['PST']=pd.to_datetime(df['PST'])
# 날짜데이터 datetime으로 변형
# df.info()
fig=plt.figure(figsize=(10,6))
plt.plot(df['PST'], df['Max TemperatureF'], c='red', alpha=0.5)
# dpi는 안먹는듯 alpha 투명도
plt.plot(df['PST'],df['Min TemperatureF'], c= 'blue', alpha= 0.5)
plt.fill_between(df['PST'],df['Max TemperatureF'],df['Min TemperatureF'],
facecolor='violet',alpha=0.2)
# 사이 채우기
plt.title("평균 기온 통계",fontsize=20)
plt.xlabel('기간',fontsize=16)
plt.ylabel('기온',fontsize=16)
plt.savefig('data/chart.png')
# 그래프파일 저장
plt.show()
```

온도 데이터를 사용하여 3행 1열 subplot 작성
1행 : 날짜-최저기온, 2행 날짜-평균기온, 3행: 날짜 최고기온


```python
filepath='data/temperature_2014.csv'
df=pd.read_csv(filepath)
df['PST']=pd.to_datetime(df['PST'])
# df.info()
plt.figure(figsize=(6,12))
plt.subplot(3,1,1)
plt.plot(df['PST'],df['Min TemperatureF'])
plt.title('최저기온',fontsize=12)
plt.subplot(3,1,2)
plt.plot(df['PST'],df['Mean TemperatureF'])
plt.title('평균기온',fontsize=12)
plt.subplot(3,1,3)
plt.plot(df['PST'],df['Max TemperatureF'])
plt.title('최고기온',fontsize=12)
plt.show()
```
