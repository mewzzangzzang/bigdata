```python
import numpy as np
import pandas as pd
import matplotlib 
import matplotlib.pylab as plt
import seaborn as sns
from matplotlib import rc, font_manager
```

```python
# 한글인코딩위한 폰트설정
fontname='c:/Windows/Fonts/malgun.ttf'
font_name=font_manager.FontProperties(fname=fontname).get_name()
rc('font', family=font_name)
```

```python
def draw_blockMap(blockedMap,targetData,title,color):
  # print('blockMap 그리기')
  # 지도 안에 라벨링(행정군구 이름을 표시) 타이틀을 어떻게 나타낼것인가(whitelabel)
  whitelabelmin=(max(blockedMap[targetData]-min(blockedMap[targetData]))*0.25+min((blockedMap[targetData])))
  # 데이터 라벨설정
  datalabel=targetData
  vmin=min(blockedMap[targetData])
  vmax=max(blockedMap[targetData])
  # 블록앱의 데이터는 기본값으로 인덱스 설정 y 엑셀데이터를 데이터프레임화 바이브(중복값x) 
  mapdata=blockedMap.pivot(index='y',columns='x',values=targetData)
  masked_mapdata=np.ma.masked_where(np.isnan(mapdata),mapdata)
  plt.figure(figsize=(8,13))
  plt.title(title)
  plt.pcolor(masked_mapdata,vmin=vmin,vmax=vmax,cmap=color,edgecolor='#aaaaaa',linewidth=0.5)
  
  # 행정구역 글
  for idx,row in blockedMap.iterrows():
    annocolor='white' if row[targetData] > whitelabelmin else 'black'
    if row['광역시도'].endswith('시') and not row['광역시도'].startswith('세종'):
      dispname='{}\n{}'.format(row['광역시도'][:2],row['행정구역'][:-1])
      if len(row['행정구역'])<=2:
        dispname+=row['행정구역'][-1]
      else:
        dispname=row['행정구역'][:-1] 
      
      if len(dispname.splitlines()[-1])>=3:
        fontsize,linespacing=9.5,1.5
      else:
        fontsize,linespacing=11,1.2
      plt.annotate(dispname,(row['x']+0.5,row['y']+0.5),weight='bold',fontsize=fontsize,ha='center',va='center',linespacing=linespacing)  
      for path in BORDER_LINES:
        ys,xs=zip(*path)

  plt.show()
```

```python
draw_blockMap(data_population,'count','행정구역별 공공보건의료기관수','Reds')
```
