# 상관관계분석과 상관계수
- 상관관계분석: 두 변수사이의 상관정도를 분석하는 것
- 상관계수 : 상관관계 강도를 나타는 것



상관계수(기호 : r)
 - -1≤r≤1 의 값을 가짐



- 상관관계 분석식(모집단 전체의 상관계수 공식)




- 상관관계 분석식(표본: 일부 샘플들의 상관계수 공식)



상관관계분석1 참고

#https://bluemumin.github.io/python/2022/06/22/Python-%EC%83%81%EA%B4%80%EA%B3%84%EC%88%98-%EA%B7%B8%EB%9E%98%ED%94%84-%EA%B7%B8%EB%A6%AC%EA%B8%B0/

```python
import math

def correlation(x,y):
    n=len(x)
    x_sum=0.0
    y_sum=0.0
    x_sum_pow=0.0
    y_sum_pow=0.0
    mul_xy_sum=0.0

    for i in range(n):
        mul_xy_sum = mul_xy_sum + float(x[i]) * float(y[i])
        x_sum = x_sum + float(x[i])
        y_sum = y_sum + float(y[i])
        x_sum_pow=x_sum_pow + pow(float(x[i]),2)
        y_sum_pow=y_sum_pow + pow(float(y[i]),2)

        try:
          r=((n*mul_xy_sum)-(x_sum*y_sum)) / math.sqrt(((n*x_sum_pow)-pow(x_sum,2)) * ((n*y_sum_pow)-pow(y_sum,2)))
        except:
          r=0.0 
    return r
```

```python
print(h)
# 키
print(height)
# 산 높이
print(foot_size)
print('키와 발크기의 대한 상관계수:',correlation(h,foot_size))
print('산의 높이와 기온사이 상관계수:',correlation(height,temper))
print('임의의 배열 x1,x2 사이의 상관계수:',correlation(x1,x2))
```

```python
df_corr_heatmap=sns.heatmap(df_corr,cbar=True,annot=True, annot_kws={'size':15},fmt='.3f',square=True,cmap='Reds',vmin=0.9,vmax=1)
# vmax,vmin
```

```python
import numpy as np
fig,ax=plt.subplots(figsize=(7,7))
mask=np.zeros_like(df_corr,dtype=np.bool_)
# 데이터 프레임의 크기만큼 0으로 채움, 데이터 타입을 bool로 고쳐주세요~
mask[np.triu_indices_from(mask)]=True
# triu_indices_from 을 주로사용 왜냐면 상위에 표현하는게 가시성이 좋아요~
# tril_indices_from 하위표출 
df_corr_heatmap=sns.heatmap(df_corr,cbar=True,mask=mask,
                            annot=True, annot_kws={'size':15},fmt='.3f',square=True,cmap='Reds',vmin=0.9,vmax=1)
# 필요없는 데이터들을 마스크
```
