```python
import pandas as pd
import numpy as np
# https://www.kaggle.com/code/prasadperera/the-boston-housing-dataset
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df=pd.read_csv(data_url,sep="\s+",skiprows=22, header=None)
raw_df
# df로부터 2간씩 짬푸하면서 가져오기/ 
```

```python
data=np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
data
```

```python
target=raw_df.values[1::2,2]
target
```

```python
df=pd.DataFrame(data,columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 
                              'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
df.head()
df['TARGET']=target
df.head()
```

요약통계
1. chas에 따른 집가격(target)위 평균, 표준편차, 중앙값을 구하라.
2. CHAS(찰스강 경계)따른 집값에 대한  t테스트 수행
3. 선형회귀분석을 모델을 생성하고 모델을 사용하여 예측값을 구하라.
4. 새로운 데이터 2개를 사용하여 예측값을 구하라.
5. 부분회귀분석 시각화를 작성하여 독립변수들이 종속변수에 미치는 영향을 분석하라. 
 - 1개독립변수 분석(plot_partregress)은 (CRIM)으로한다.
 - plot_partregress_grid() 작성

```python
df.describe()
```

```python
# chas에 따른 집가격(target)위 평균, 표준편차, 중앙값을 구하라.
df.groupby(target)['CHAS'].agg(['mean','std'])
```

```python
from scipy import stats
from statsmodels.formula.api import ols, glm
```

```python
# CHAS(찰스강 경계)따른 집값에 대한  t테스트 수행
df_1=df.loc[df['CHAS']==1,'TARGET']
df_0=df.loc[df['CHAS']==0,'TARGET']
# 확인
print(df_1[10:20])
print(df_0[10:20])
```

```python
# ttest
stats.ttest_ind(df_1,df_0,equal_var=False)
```

```python
# 선형회귀분석을 모델을 생성하고 모델을 사용하여 예측값을 구하라.
Rformula = 'TARGET ~ CRIM + ZN + INDUS + \
       NOX + RM + AGE + \
      DIS + TAX + PTRATIO + B + LSTAT' 
```

```python
regression_result=ols(Rformula,data=df).fit()
```

```python
regression_result.summary()
```

```python
# 새로운 데이터 2개를 사용하여 예측값을 구하라.
sample1=df[df.columns.difference(['CHAS','TARGET'])]
sample1.head()
```

```python
sample1_predict=regression_result.predict(sample1)
```

```python
print(sample1_predict[:10])
```
