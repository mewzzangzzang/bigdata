```python
# 좌표 100개 생성
import numpy as np
from folium import plugins
n=100
data=np.array([np.random.uniform(low=35,high=38, size=n),np.random.uniform(low=126,high=129, size=n)]).T
data
# pop은 문자열로 저장되어짐
popup=[str(i) for i in range(1, n+1)]
# 지도 줌레벨에 따라서  합산수 산출 최대시 개별
m=folium.Map(location=[latitude,longitude],zoom_start=7)
# 플러그인 사용해서 데이터좌표 수
plugins.MarkerCluster(data, popups=popup).add_to(m)
m
```

```python
import pandas as pd
df=pd.read_csv('data/CCTV_20190917.csv')
df
# 카메라 대수 0인 데이터 삭제 (위도, 경도, 대수)필요
```

```python
# 카메라 대수 0인 데이터 삭제 (위도, 경도, 대수)필요
# 데이터 정제(축소)
df1=df.loc[df['카메라대수']>0,['소재지도로명주소','위도','경도']]
df1
```

```python
# 팝업 리스트
# 리스트화 해야함
popups=df1['소재지도로명주소'].to_list()
popups
# 위도경도 리스트화
lat=df1['위도'].to_list()
# lat
lng=df1['경도'].to_list()
# lng

# 리스트는 평균x np사용해서 배열화, 평균구하기
data=np.array([lat,lng]).T
# 평균 좌표를 기준점으로 map zoom
lat_avg=np.array(lat).mean()
lng_avg=np.array(lng).mean()
print(lat_avg,lng_avg)

m=folium.Map(location=[lat_avg,lng_avg],zoom_start=14)
plugins.MarkerCluster(data,popups=popups).add_to(m)
m
```
