```python
addr['시도'].unique()
```

```python
addr_aliases={'경기':'경기도','경남':'경상남도','경북':'경상북도',
              '충북':'충청북도','충남':'충청남도','전남':'전라남도',
              '서울시':'서울특별시','부산특별시':'부산광역시',
              '대전시':'대전광역시','강원도':'강원특별자치도',
              '전북':'전라북도'}
```

```python
addr['시도']=addr['시도'].apply(lambda v: addr_aliases.get(v,v))
addr['시도'].unique()
```

```python
addr['군구'].unique()
```

```python
addr[addr['군구']=='아란13길']
addr.iloc[75]=['제주특별자치구','제주시']
```

```python
addr[addr['군구']=='동매로']
addr.iloc[155]=['부산광역시','사하구']
```

```python
addr['군구'].unique()
```

```python
addr['시도군구']=addr.apply(lambda r:r['시도']+' '+r['군구'], axis=1)
addr['count']=0
addr
```

```python
addr_groupby=pd.DataFrame(addr.groupby(['시도','군구','시도군구'], 
                                       as_index=False).count())
addr_groupby=addr_groupby.set_index('시도군구')
addr_groupby
```

```python
population=pd.read_excel('data/행정구역_시군구_별__성별_인구수_20230907141126.xlsx')
population
```

```python
population['시도'].unique()
```

```python
population['군구']=population['군구'].apply(lambda v:v.strip())
population['군구'].unique()
```

```python
population['시도군구']=population.apply(lambda r:r['시도']+' '+r['군구'], axis=1)
population=population[population['군구']!='전국']
population=population[population['군구']!='소계']
population=population.set_index('시도군구')
population
```

```python
addr_polulation=pd.merge(addr_groupby,population, how='inner')
addr_polulation
```
