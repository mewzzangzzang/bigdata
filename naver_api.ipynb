```python
# 네이버 검색 API 예제 - 블로그 검색
import os 
import sys
import urllib.request
import json
client_id='9bLbOBlav89U3DdVt3iO'
client_secret='VGEXL3fXtT'
encText = urllib.parse.quote("다이어트")
url = "https://openapi.naver.com/v1/search/news?query=" + encText + "&display=100"
# 100가지 추출
# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
# 네이버에서는 헤더필요
response = urllib.request.urlopen(request)
# 요청이 정상적이라면 
rescode = response.getcode()
img_link_list=[]

if(rescode==200):
    response_body = response.read()
    res_body=json.loads(response_body.decode('utf-8'))
    #print(response_body.decode('utf-8'))
    for item in res_body['items']:
        img_link_list.append(item['link'])
    print(img_link_list)
else:
    print("Error Code:" + rescode)
```

```python
i=1
for link in img_link_list:
       save_name='data/images/img'+str(i)+'.jpg'
       
       try:
        urllib.request.urlretrieve(link, save_name)
        i+=1
       except:
           print("Error")
```
