```python
import numpy as np
import pandas as pd
import matplotlib 
import matplotlib.pylab as plt
import seaborn as sns
from matplotlib import rc, font_manager
```

```python
sns.set(style='whitegrid') #스타일 지정
# 한글인코딩위한 폰트설정
fontpath='c:/Windows/Fonts/malgun.ttf'
font_name=font_manager.FontProperties(fname=fontpath).get_name()
rc('font', family=font_name)

# -음수 자동 유니코드  설정
plt.rcParams['axes.unicode_minus']=False
# 차트 피규어 크기 설정
plt.rcParams['figure.figsize']=(3,3)
```

```python
# 상관관계
titanic=sns.load_dataset('titanic')
corr_titanic=titanic.corr()
corr_titanic
```

```python
sns.pairplot(titanic,hue='survived',vars=['parch','fare'])
```

```python
fig,ax=plt.subplots(figsize=(9,9))
corr_titanic_heapmap=sns.heatmap(corr_titanic,annot=True)
```

```python
import numpy as np
# 상위
fig,ax=plt.subplots(figsize=(9,9))
mask=np.zeros_like(corr_titanic,dtype=np.bool_)
mask[np.triu_indices_from(mask)]=True
corr_titanic_heapmap=sns.heatmap(corr_titanic,cbar=True, mask=mask,
                            annot=True, annot_kws={'size':15},fmt='.2f',square=True,cmap='Blues')
```

```python
# 최소값
import numpy as np
fig,ax=plt.subplots(figsize=(9,9))
mask=np.zeros_like(corr_titanic,dtype=np.bool_)
mask[np.triu_indices_from(mask)]=True
corr_titanic_heapmap=sns.heatmap(corr_titanic,cbar=True, mask=mask,
                            annot=True, annot_kws={'size':15},fmt='.2f',square=True,cmap='Blues',vmin='0.0')
```
